```python
#그룹화
def group_df(df):
  df_gb = df.groupby("연관어")["건수"].sum().reset_index()
  df_gb = df_gb.sort_values("건수", ascending=False).reset_index(drop=True)
  return df_gb

def word_cloud(df, year=1):
  wc = df.set_index("연관어").to_dict()["건수"]
  wordCloud = WordCloud(font_path = font,
                        width = 400, # 워드 클라우드의 너비 지정
                        height = 400, # 워드클라우드의 높이 지정
                        max_font_size=100, # 가장 빈도수가 높은 단어의 폰트 사이즈 지정
                        background_color = 'white' # 배경색 지정
                        ).generate_from_frequencies(wc) # 워드 클라우드 빈도수 지정
  plt.figure(figsize=(10,10))
  plt.imshow(wordCloud)
  plt.axis("off")

  plt.savefig(f"공모전/행정안전부")
```

```python
from nltk import word_tokenize, bigrams #토큰화, 바이그램

from nltk.util import ngrams #트라이그램
from nltk import ConditionalFreqDist #동시출현빈도 계수화


import gensim
from gensim.models import CoherenceModel
from gensim import corpora
from gensim.parsing.preprocessing import preprocess_string
```

```python
word_list = []
for word in df2["키워드"]:
  word_list.append(word.split(" "))

bigram = gensim.models.Phrases(word_list)
# trigram = gensim.models.Phrases(bigram[word_list])
bigram_mod = gensim.models.phrases.Phraser(bigram)
# trigram_mod = gensim.models.phrases.Phraser(trigram)

bigram_document = [bigram_mod[doc] for doc in word_list]
id2word = corpora.Dictionary(bigram_document)
corpus = [id2word.doc2bow(doc) for doc in bigram_document]

coherence_values = []
for i in range(2, 11):
  ldamodel = gensim.models.ldamodel.LdaModel(corpus, id2word=id2word, num_topics=i, random_state=17)
  coherence_model_lda = CoherenceModel(model = ldamodel, texts=bigram_document, dictionary=id2word, topn=10)
  coherence_lda = coherence_model_lda.get_coherence()
  coherence_values.append(coherence_lda)
  print('n:', i, '\nCoherence Score:', coherence_lda)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("fivethirtyeight")

x = range(2, 11)

plt.figure(figsize=(15, 7))
plt.plot(x, coherence_values)
plt.title('Coherence Values', size=25)
plt.xlabel('Number of Topics')
plt.xlim(1, 10)
plt.xticks(x, range(2, 10), fontsize=15)

plt.ylabel('Coherence Score')
plt.yticks(fontsize=15)
# plt.grid()
plt.show()
```

```python
class CustomTokenizer:
  def __init__(self, tagger):
    self.tagger = tagger
  def __call__(self, sent):
    word_tokens = self.tagger.morphs(sent)
    result = [word for word in word_tokens if len(word) > 1]
    return result
```

```python
df2 = df2.reset_index(drop=True)
df3 = df2
print("df3 length:", len(df3))

years = df3.get("연도")
posts = df3.get("키워드")

nouns1 = df3["키워드"].values.tolist()
nouns2 = []
for sentence in nouns1:
  # print(sentence.split(" "))
  nouns2.append(sentence.split(" "))
print("nouns2 length(df3 length와 같아야함)", len(nouns2))

# keyword_list = df2["키워드"].values.tolist()
# print(len(nouns2))

nouns2_year = {}
for i in range(len(posts)): #0~980
  if years[i] in nouns2_year:
    nouns2_year[years[i]].append(nouns2[i])
  else:
    nouns2_year[years[i]] = []
    nouns2_year[years[i]].append(nouns2[i])
        
nouns2_year = sorted(nouns2_year.items())        

print(len(nouns2_year))
```

```python
#연도별 논문개수
nouns2_years = [] #해당연도에 해당하는 단어들
years_length = [] #해당연도에 기사가 몇개 있는지

for i in range(len(nouns2_year)):
  nouns2_years.append(nouns2_year[i][1])
  years_length.append(len(nouns2_year[i][1]))

print(len(nouns2_years))
print(years_length)
print("연도별 기사 갯수의 합(df length와 같아야함):", sum(years_length))
```

```python
#연도별 명사들이 있는 데이터 하나로 통합
merged_years = []
for texts in nouns2_years:
  for text in texts:
    merged_years = merged_years + [text]
```

```python
nouns2_topics = []
for i in range(len(ctm.get_topics(20))):
  nouns2_topics.append(ctm.get_topics(20)[i])
len(nouns2_topics)
```

```python
#날짜별로 합쳐진 문서마다 토픽들의 크기 계산
nouns2_topic_size=[]
index=0

for post_len in years_length: #연도별 논문개수 리스트에서 추출[4, 1, 1, ,,,,,]
  topic_size=[]
  doc=[]
  print(post_len) #4, 1, 1, ,,,,
  print('----------- 하위---------')
  
  for i in range(post_len): #i = 4, 1, 1, ,,,,,
    doc+=merged_years[index]
    index+=1
      
  # print(doc, '\n')
  
  for model in nouns2_topics: #[['운동', '프로그램', '신체_활동',,,,,,]]
    count=0
    for noun in doc:
      if noun in model:
        count+=1
    topic_size.append(count)
  nouns2_topic_size.append(topic_size)
    
# print(nouns2_topic_size)
```

```python
nouns2_topic_array = np.array(nouns2_topic_size)
nouns2_topic_df = pd.DataFrame(nouns2_topic_array, 
                              index = sorted(list(set(years))),
                              columns = ['topic1', 'topic2', 'topic3', 'topic4', 'topic5', 'topic6', "topic7", "topic8", "topic9"])
nouns2_topic_df.head()
```
